```python
import pandas as pd
import numpy as np
df = pd.read_csv('C:/Users/user/Store_Data.csv')

df
```

```python
# let's calculate the tax rate

tax_rate = np.array([0.15, 0.20, 0.25])

df['tax_' + str(tax_rate[0])] = df['new_net_revenue_per_product_with_30%_Discount'] * tax_rate[0]

df['tax_' + str(tax_rate[1])] = df['new_net_revenue_per_product_with_30%_Discount'] * tax_rate[1]

df['tax_' + str(tax_rate[2])] = df['new_net_revenue_per_product_with_30%_Discount'] * tax_rate[2]
```

```python
df
```

```python
# now this is the new net revenue per product with 30% discount after these tax rate


df['new_net_revenue_per_product_with_30%_Discount_after_tax_rate_15%'] = df['new_net_revenue_per_product_with_30%_Discount'] - df['tax_' + str(tax_rate[0])]

df['new_net_revenue_per_product_with_30%_Discount_after_tax_rate_20%'] = df['new_net_revenue_per_product_with_30%_Discount'] - df['tax_' + str(tax_rate[1])]

df['new_net_revenue_per_product_with_30%_Discount_after_tax_rate_25%'] = df['new_net_revenue_per_product_with_30%_Discount'] - df['tax_' + str(tax_rate[2])]
```
